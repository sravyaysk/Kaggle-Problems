```python
import pandas as pd
import math
import os
import csv
```

```python
train_df = pd.read_csv("E:/IIITH AIML/week14/Train_Kaggle.csv")
test_df = pd.read_csv('E:/IIITH AIML/week14/Test_Kaggle.csv')
combine = [train_df, test_df]
```

```python
train_data = train_df.values
sales = []
for i in train_data:
    if(math.isnan(i[3])):
        sales.append(0)
    else:
        sales.append(i[3])
```

```python
#replacing nan values
replace_value = sum(sales)/len(sales)
for i in train_data:
    if(math.isnan(i[3])):
        i[3] = replace_value
```

```python
for dataset in combine:
    dataset['ProductCategory'] = dataset['ProductCategory'].map( {'WomenClothing': 1, 'MenClothing': 2, 'OtherClothing': 3} ).astype(int)
```
