```python
train_df = pd.read_csv("E:/IIITH AIML/week14/Train_Kaggle.csv")
test_df = pd.read_csv('E:/IIITH AIML/week14/Test_Kaggle.csv')
xlsx1 = pd.ExcelFile('E:/IIITH AIML/week14/MacroEconomicData.xlsx')
dataset1 = pd.read_excel(xlsx1, 'Sheet1')
xlsx2 = pd.ExcelFile('E:/IIITH AIML/week14/WeatherData.xlsx')
dataset21 = pd.read_excel(xlsx2, '2009')
dataset22 = pd.read_excel(xlsx2, '2010')
dataset23 = pd.read_excel(xlsx2, '2011')
dataset24 = pd.read_excel(xlsx2, '2012')
dataset25 = pd.read_excel(xlsx2, '2013')
dataset26 = pd.read_excel(xlsx2, '2014')
dataset27 = pd.read_excel(xlsx2, '2015')
dataset28 = pd.read_excel(xlsx2, '2016')
xlsx3 = pd.ExcelFile('E:/IIITH AIML/week14/Events_HolidaysData.xlsx')
dataset3 = pd.read_excel(xlsx3, 'Sheet1')
combine = [train_df, test_df]
```

```python
file_names = ["new_train_data1","new_test_data1"]
dependency_1_values = dataset1.values
```

```python
#combining train_df(test_df) and dataset1
month_dict = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
headers = ['Year','Month','ProductCategory','Sales(In ThousandDollars)','Year-Month','Monthly Nominal GDP Index (inMillion$)',
           'Monthly Real GDP Index (inMillion$)','CPI','PartyInPower','unemployment rate','CommercialBankInterestRateonCreditCardPlans',
           'Finance Rate on Personal Loans at Commercial Banks, 24 Month Loan','Earnings or wages  in dollars per hour',
           'AdvertisingExpenses (in Thousand Dollars)','Cotton Monthly Price - US cents per Pound(lbs)','Change(in%)',
           'Average upland planted(million acres)','Average upland harvested(million acres)','yieldperharvested acre',
           'Production (in  480-lb netweright in million bales)','Mill use  (in  480-lb netweright in million bales)','Exports']
index = 0
j = 0
for dataset in combine:
    #print(dataset)
    data = dataset.values
    i = 0
    output_filename = os.path.join('E:/IIITH AIML/week14/'+file_names[index]+".csv")
    new_data = []
    new_data.append(headers)
    while(i < len(data)):
        inner = []
        temp_arr = (dependency_1_values[j][0].split(' '))
        if( (int(temp_arr[0]) == int(data[i][0])) and (int(data[i][1]) == int(month_dict[temp_arr[2]])) ):
            for l in range(0,4):
                inner.append(data[i][l])
            for k in range(0,18):
                inner.append(dependency_1_values[j][k])
        if(i%3 == 2):
            j = j + 1
        i = i + 1
        new_data.append(inner)
    f = open(output_filename, 'ab')
    writer = csv.writer(f)
    for i in new_data:
        writer.writerow(i)
    f.close()
    index = index + 1
```
